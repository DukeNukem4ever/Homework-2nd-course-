```python
import json
import urllib.request

lisst = []
lisst2 = []
languages = []
repos = []

url01 = 'https://api.github.com/users/elmiram/repos?page=1&per_page=200'
response1 = urllib.request.urlopen(url01)
text01 = response1.read().decode('utf-8')
data01 = json.loads(text01)

url02 = 'https://api.github.com/users/maryszmary/repos?page=1&per_page=200'
response02 = urllib.request.urlopen(url02)
text02 = response02.read().decode('utf-8')
data02 = json.loads(text02)

url03 = 'https://api.github.com/users/lizaku/repos?page=1&per_page=200'
response03 = urllib.request.urlopen(url03)
text03 = response03.read().decode('utf-8')
data03 = json.loads(text03)

url04 = 'https://api.github.com/users/nevmenandr/repos?page=1&per_page=200' 
response04 = urllib.request.urlopen(url04)
text04 = response04.read().decode('utf-8')
data04 = json.loads(text04)

url05 = 'https://api.github.com/users/ancatmara/repos?page=1&per_page=200'
response05 = urllib.request.urlopen(url05)
text05 = response05.read().decode('utf-8')
data05 = json.loads(text05)


for i in data01:
    languages.append(i["language"])

for i in data02:
    languages.append(i["language"])

for i in data03:
    languages.append(i["language"])

for i in data04:
    languages.append(i["language"])

for i in data05:
    languages.append(i["language"])
    
repos.append(str(len(data01)))
repos.append(str(len(data02)))
repos.append(str(len(data03)))
repos.append(str(len(data04)))
repos.append(str(len(data05)))

def most_common(l):
    max = 0
    maxitem = None
    for x in set(l):
        count =  l.count(x)
        if count > max:
            max = count
            maxitem = x
    return maxitem

user = input("Введите одного из данных пользователей: elmiram, maryszmary, lizaku, nevmenandr, ancatmara. \n")

while user != "elmiram" and user!= "maryszmary" and user != "lizaku" and user != "nevmenandr" and user != "ancatmara":
    print("Пожалуйста, введите один из указанных адресов пользователя!")
    user = input("Введите одного из данных пользователей: elmiram, maryszmary, lizaku, nevmenandr, ancatmara \n")

def data(user):
    url = 'https://api.github.com/users/%s/repos?page=1&per_page=200' % user
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')
    return json.loads(text)

def data2(user):
    followers = 'https://api.github.com/users/%s/followers?page=1&per_page=200' % user
    follows = urllib.request.urlopen(followers)
    text2 = follows.read().decode('utf-8')
    return json.loads(text2)

print()
print("Всего репозиториев: " + str(len(data(user))))
print("Все репозитории: ")
for i in data(user):
    print(i["name"], " - ", i["description"])
print()
print("Все языки программирования, на которых пишет пользователь, и репозитории, на которых они написаны:")
print()    
for i in data(user):
    lisst2.append(i["language"])
    if i["language"] not in lisst:
            lisst.append(i["language"])
    print(i["language"], " - ", i["name"])
if None in lisst:
    lisst.remove(None)
if None in lisst2:
    lisst2.remove(None)
print()
print("Главные языки программирования, на которых пишет пользователь " + user + ": ")
print()
for word in lisst:
    print(word)
print()
print("Самый популярный язык, на котором пишет пользователь " + user + ": ")
print()
print(most_common(lisst2))
print()
print("Количество последователей: " + str(len(data2(user))))
print()
print("Самое большое количество репозиториев у пользователя elmiram.")
print()
print()
print("Самый популярный язык среди большинства пользователей списка: " + most_common(languages)) 
print()
print("Самое большое количество подписчиков у пользователя ancatmara.")
```
